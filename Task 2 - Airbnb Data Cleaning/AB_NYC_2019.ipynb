```python
df["name"] = df["name"].fillna("Unknown")
```

```python
df["host_name"] = df["host_name"].fillna("Unknown")
```

```python
df["last_review"] = df["last_review"].fillna(pd.Timestamp("1900-01-01"))
```

```python
print(df.isnull().sum())
```

```python
Q1 = df["price"].quantile(0.25)
Q3 = df["price"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[
    (df["price"] < lower) |
    (df["price"] > upper)
]

print("Number of Outliers:", len(outliers))
```

```python
df = df[
    (df["price"] >= lower) &
    (df["price"] <= upper)
]
```

```python
print(df.shape)
```

```python
df = df.copy()
```

```python
df = df[
    (df["price"] >= lower) &
    (df["price"] <= upper)
].copy()
```

```python
df["host_name"] = df["host_name"].str.strip()
df["neighbourhood_group"] = df["neighbourhood_group"].str.strip()
df["room_type"] = df["room_type"].str.strip()
```

```python
df.loc[:, "host_name"] = df["host_name"].str.strip()
df.loc[:, "neighbourhood_group"] = df["neighbourhood_group"].str.strip()
df.loc[:, "room_type"] = df["room_type"].str.strip()
```
